```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-13, 13, 1000)

handle1, = plt.plot(x, 1/np.sin(x) + np.cos(x), '-m' ,label = 'y = 1/sin(x) + cos(x)')
handle2, = plt.plot(x, (1/x ** 2) * np.e ** ((-1)/x ** 2), '--c',label = 'y = (1/x ** 2) * np.e ** ((-1)/x ** 2)')
plt.legend(handles = [handle1, handle2])

plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2.0*np.pi, 2.0*np.pi, 100)

handle1, = plt.plot(x, np.sin(x), ':r',label = 'y = sin(x)')
handle2, = plt.plot(x, np.sin(x) - 2,'-.b', label = 'y = sin(x) - 2')
handle3, = plt.plot(x, np.sin(x) + 1, '--g', label = 'y = sin(x) + 1')

plt.axhline(y = 0, color = 'k') 
plt.axvline(x = 0, color = 'k')

plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2.0*np.pi, 2.0*np.pi, 100)

handle1, = plt.plot(x, np.cos(x), ':r',label = 'y = cos(x)')
handle2, = plt.plot(x, 2 * np.cos(x),'-.b', label = 'y = 2cos(x)')
handle3, = plt.plot(x, 0.3 * np.cos(x), '--g', label = 'y = 0.3cos(x)')
handle4, = plt.plot(x, (-1) * np.cos(x), '-c', label = 'y = -cos(x)')

plt.legend(handles = [handle1, handle2, handle3, handle4])

plt.axhline(y = 0, color = 'k') 
plt.axvline(x = 0, color = 'k')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5,100)

def func(k):
    return abs(abs(k) - 2)

plt.subplot(2,3,1)
plt.plot(x, func(x))
plt.grid(True)

plt.subplot(2,3,2)
plt.plot(x, func(x - 2), 'r')
plt.grid(True)

plt.subplot(2,3,3)
plt.plot(x, func(x + 2), 'g')
plt.grid(True)

plt.subplot(2,3,4)
plt.plot(x, func(2 * x), 'c')
plt.grid(True)

plt.subplot(2,3,5)
plt.plot(x, func(0.5 * x), 'm')
plt.grid(True)

plt.subplot(2,3,6)
plt.plot(x, func(-x), 'y')
plt.grid(True)
```

каждый из графиков обращается к созданной функции func(k) и возвращает график в соответствие с abs(abs(k) - 2)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3, 4, 1)

z = np.ones((1, 7))*10

y = np.power(z, x)

plt.subplot(2, 1, 1)
plt.plot(x, y[0])

plt.subplot(2, 1, 2)
plt.semilogy(x, y[0])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi, 100)

handle1, = plt.plot(x, np.cos(x), '--g', label = 'y = cos(x)')
handle2, = plt.plot(x, np.arccos(x), '-.r', label = 'y = arccos(x)')
handle3, = plt.plot(x, x, label = 'y = x')

plt.legend(handles=[handle1, handle2, handle3])

plt.axhline(y = 0, color = 'k') 
plt.axvline(x = 0, color = 'k')

plt.xlabel('x')
plt.ylabel('y')

plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 5, 100)

plt.subplot(2, 1, 1)
plt.plot(x, (x + 3) ** 0.5, '-b')
plt.title('$y = sqrt(x+3)$')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, np.sign(x),'-.y')
plt.title('$y = sign(x)$')
plt.grid(True)
plt.show()
```
