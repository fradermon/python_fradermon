```python
import numpy as np

q = complex(256, 0)
w = np.abs(q)
e = np.arange(1, 9)
r = np.angle(q)

root = w * (1 / 8) * (np.cos(r + np.pi * e / 4) - 1j * np.sin(r + np.pi * e / 4))

for number in root:
    if number < 0:
        print(number)
```

```python
root_r = np.around(root, 2)
for number in root_r:
    if np.imag(number) == 0:
        print(number)
```

```python
for number in root:
    if np.angle(number) > np.pi / 3:
        print(number)
```
