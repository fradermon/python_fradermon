```python
import matplotlib.pyplot as plt
import numpy as np

z1, z2, z3, z4 = 1 + 3j, 3 + 4j, (z1 + z2), (z1 - z2)
x = np.array([z1, z2, z3, z4])
orig = np.zeros(4)

plt.plot()

q = plt.quiver(orig,orig, np.real(z1), np.imag(z1),
angles='xy', scale_units='xy', scale=1, color="green")
q = plt.quiver(orig,orig, np.real(z2), np.imag(z2),
angles='xy', scale_units='xy', scale=1, color="blue")
q = plt.quiver(orig,orig, np.real(z3), np.imag(z3),
angles='xy', scale_units='xy', scale=1, color="violet")
q = plt.quiver(orig,orig, np.real(z4), np.imag(z4),
angles='xy', scale_units='xy', scale=1, color="red")

plt.axis([-8,8,-8,8])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.grid(True)

plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

z = 2 * np.sqrt(3) - 2j

r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,7)

zroot = r ** (1/4) * ( np.cos((phi + 2.0 * np.pi * k) / 6) + 1j * np.sin((phi + 2.0 * np.pi * k) / 6))

plt.title('$\sqrt[4]{2\sqrt{3}-2j}$ complex roots')
plt.plot(np.real(zroot), np.imag(zroot), '*g')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('equal')

t = np.linspace(0,2*np.pi,100)

plt.plot(r ** (1 / 4) * np.cos(t),r ** (1 / 4) * np.sin(t),':')

plt.plot([np.real(zroot)[0], np.real(zroot)[1], np.real(zroot)[2], np.real(zroot)[3], np.real(zroot)[4], np.real(zroot)[5], np.real(zroot)[0]],\
         [np.imag(zroot)[0], np.imag(zroot)[1], np.imag(zroot)[2], np.imag(zroot)[3], np.imag(zroot)[4], np.imag(zroot)[5], np.imag(zroot)[0]],'-.',color='red')

plt.xlabel('Real')
plt.ylabel('Imag')

plt.show()
```
